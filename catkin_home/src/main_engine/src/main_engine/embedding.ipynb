```python
newCommand = "Go to the kitchen"
```

```python
command_embedding = get_embedding(newCommand)
```

```python
cosine_similarity = lambda x, y: np.dot(x, y)/(np.linalg.norm(x)*np.linalg.norm(y))
```

```python
nearest = 0
nearest_command = ""
for emb, command in zip([GoTo_embedding, Find_embedding, FindPerson_embedding, GotoBring_embedding], [GoTo, Find, FindPerson, GotoBring]):
    similarity = cosine_similarity(emb, command_embedding)
    if similarity > nearest:
        nearest = similarity
        nearest_command = command
```
